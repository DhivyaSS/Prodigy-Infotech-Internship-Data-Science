```python
p=df[1:18]
fig,axes=plt.subplots(1,2, figsize=(15,15))
sns.barplot(data=p,x='Male',y='State/UT',ax=axes[0])
sns.barplot(data=p,x='Female',y='State/UT',ax=axes[1])
plt.title('Population distribution of Male and Female')
plt.tight_layout()
```
