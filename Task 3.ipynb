```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
catcols=['marital','education','housing loan','personal loan','contact','job','pcoutcome','client sub term deposit']
df[catcols]=df[catcols].apply(le.fit_transform)
df[catcols]
```

```python
catcols1=['contact','contact month','default credit']
df[catcols1]=df[catcols1].apply(le.fit_transform)
df[catcols1]
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='gini',random_state=0,splitter='best')
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy = accuracy_score (y_test, y_pred) * 100
print ("Accuracy of the model is {:.2f}". format(accuracy))
```

```python
from sklearn. metrics import classification_report
class_report = classification_report (y_test, y_pred)
print (f"\nClassification Report:\n{class_report}")
```

```python
confusion_matrix(y_test,y_pred)
```
